# **Titanic Comprehensive Survival Analysis**

### What is the distribution of passengers by class?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

titanic = sns.load_dataset('titanic')

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='pclass', data=titanic, hue='pclass', palette='viridis', legend=False)
plt.title('Distribution of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

### What is the age distribution of passengers?

```python
plt.figure(figsize=(8, 6))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

### How does age vary across passenger classes?


```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='pclass', y='age', hue='pclass', data=titanic, palette='viridis', dodge=False)
plt.title('Age Distribution Across Passenger Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend([],[], frameon=False)
plt.show()
```

### What percentage of passengers survived, by class or gender?

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.barplot(x='pclass', y='survived', data=titanic, hue='pclass', palette='viridis', errorbar=None)
plt.title('Survival Rate by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.legend([],[], frameon=False)

plt.subplot(1, 2, 2)
sns.barplot(x='sex', y='survived', data=titanic, hue='sex', palette='muted', errorbar=None)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.legend([],[], frameon=False)
plt.tight_layout()
plt.show()
```

### What is the survival distribution across classes, separated by gender?

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(x='pclass', hue='survived', data=titanic[titanic['sex'] == 'male'], palette='Set1')
plt.title('Survival Distribution Across Classes (Male)')
plt.xlabel('Passenger Class')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.countplot(x='pclass', hue='survived', data=titanic[titanic['sex'] == 'female'], palette='Set1')
plt.title('Survival Distribution Across Classes (Female)')
plt.xlabel('Passenger Class')
plt.ylabel('Count')

plt.suptitle('Survival Distribution Across Classes by Gender', y=1.05)
plt.tight_layout()
plt.show()
```

### Who survived, grouped by gender?


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set1')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

### Whats the fare distribution for different classes, split by survival?

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='pclass', y='fare', hue='survived', data=titanic, palette='Set1')
plt.title('Ticket Fare Distribution by Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Ticket Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

### How does family size (SibSp -Siblings/Spouse or Parch - Parent/Children) differ across survival outcomes, by gender?


```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='survived', y='sibsp', hue='sex', data=titanic, palette='Set1')
plt.title('SibSp (Siblings/Spouse) vs Survival and Gender')
plt.xlabel('Survival (0 = No, 1 = Yes)')
plt.ylabel('SibSp (Siblings/Spouse)')

plt.subplot(1, 2, 2)
sns.boxplot(x='survived', y='parch', hue='sex', data=titanic, palette='Set1')
plt.title('Parch (Parent/Children) vs Survival and Gender')
plt.xlabel('Survival (0 = No, 1 = Yes)')
plt.ylabel('Parch (Parent/Children)')

plt.tight_layout()
plt.suptitle('SibSp and Parch Distribution by Survival and Gender', y=1.05)
plt.show()
```
